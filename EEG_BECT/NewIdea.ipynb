```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
path = './NewcsvData'
filelist = os.listdir(path)
```

```python
filepath = os.path.join(path, filelist[25])
```

```python
raw_data = pd.read_csv(filepath, sep='\t')
```

```python
raw_data[416000:418000].plot(figsize=[20,2])
```

```python
data = raw_data.values
```

```python
def window_slide(x, window, stride):
    n = int(len(x)-(window-1))
    out = np.zeros((n, window))
    for i in range(window-1):
        out[:,i] = x[i:-(window-1-i)].T
    output = out[0:-1:stride,:]
    return output
```

```python
datas = window_slide(x=data, window=80, stride=10)
datas.shape
```

```python
def plot_sub_window(data, group_num, boxsize):
    n = boxsize
    groups = n*n
    plot_data = data[(group_num-1)*groups:group_num*groups,:]
    plt.figure(figsize=[15,15])
    for k in range(plot_data.shape[0]):
        plt.subplot(n, n, k+1)
        plt.plot(plot_data[k,:])
```

```python
plot_sub_window(datas, group_num=4, boxsize=5)
```
